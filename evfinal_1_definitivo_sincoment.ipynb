```python
class TiendaOnline:  
    def __init__(self, nombre): 
        self.nombre = nombre   
        self.inventario = [] 
        self.ventas_totales = 0.0 

    def agregar_producto(self, nombre, precio, cantidad):
        nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad} 
        for producto in self.inventario: 
            if producto["nombre"].lower() == nombre.lower():  
                producto["cantidad"] += cantidad
                print(f"El producto {nombre.capitalize()} ya consta en nuestro inventario. Se han añadido {cantidad} unidades.")
                print(f"El total actual es de: {producto['cantidad']}")
                return 
        self.inventario.append(nuevo_producto)  
        print(f"Se ha añadido el producto {nombre.capitalize()} al inventario y hay un total de {cantidad}.")  

    def ver_inventario(self):
        print("El inventario es el siguiente:") 
        print("-" * 55)
        for producto in self.inventario: 
            detalles = [] 
            for key,value in producto.items():  
                if key == "precio": 
                    detalles.append(f"{key.capitalize()}: {value}€")
                else:  
                    detalles.append(f"{key.capitalize()}: {value}") 
            print("-" * 55)
            print(", ".join(detalles))  
                   
    def buscar_producto(self, nombre): 
        for b in self.inventario: 
            if b["nombre"] == nombre:
                print("Los detalles del producto buscado son los siguientes:")
                print(f"Nombre: {b['nombre'].capitalize()}, Precio: {b['precio']}€, Cantidad: {b['cantidad']}")   
                return 
        print("El producto indicado no está en el inventario") 

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:  
            if producto["nombre"] == nombre: 
                producto["cantidad"] += cantidad  
                print(f"El stock del producto {nombre.capitalize()} ha sido actualizado") 
                print(f"El total actual es de: {producto['cantidad']}")
                return  
        print("El producto indicado no está en el inventario. Por favor, primero actualiza el inventario e inténtalo de nuevo") 
    
    def eliminar_producto(self, nombre):
        for producto in self.inventario: 
            if producto["nombre"] == nombre:  
                self.inventario.remove(producto) 
                print(f"El siguiente producto ha sido eliminado: {nombre.capitalize()}")
                return  
        print("El producto indicado no está en el inventario, por favor revisa el nombre del producto e inténtalo otra vez ") 

    def calcular_valor_inventario(self):
        print("El valor del inventario es el siguiente:") 
        print("-" * 40)
        valor_inventario = 0 
        for producto in self.inventario: 
            total_por_producto = producto["precio"] * producto["cantidad"] 
            print(f"{producto['nombre'].capitalize()}: {producto['cantidad']} unidades x {producto['precio']}€ = {total_por_producto}€")
            print("-" * 40)
            valor_inventario += total_por_producto 

        print(f"El VALOR TOTAL DEL INVENTARIO es: {valor_inventario}€")          
```
