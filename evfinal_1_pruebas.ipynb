```python
class TiendaOnline: #definimos la clase 
    def __init__(self, nombre): #arrancamos el constructor 
        self.nombre = nombre #añado la variable de nombre para poder ponerle un nombre a la tienda 
        self.inventario = [] #añado una lista vacía que será de diccionarios 
        self.ventas_totales = 0.0 #empezamos con 0 ventas totales, con este formato porque es un float

    def agregar_producto(self, nombre, precio, cantidad):
        nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad} #añado la estructura que tiene que seguir para añadir un nuevo producto 
        for producto in self.inventario: #comienzo un bucle for para que itere el diccionario con el inventario a ver si el producto está ya en stock o no 
            if producto["nombre"] == nombre: #si el nombre introducido del producto ya está en el inventario, le digo que sume la nueva cantidad a la ya existente
                producto["cantidad"] += cantidad
                print(f"El producto {nombre} ya consta en nuestro inventario. Se han añadido un total de {cantidad} unidades.")
                return #termino el bucle para que si ese es el caso pare
        nuevo_producto not in self.inventario #si ha terminado de iterar el diccionario y no está el producto 
        self.inventario.append(nuevo_producto) #uso la función append para que añada el nuevo producto con la estructura de la variable que he añadido arriba 
        print(f"Se ha añadido el producto {nombre} al inventario y hay un total de {cantidad}.") #además especifica cuál es el nuevo producto y cuanta cantidad hay de este 

    def ver_inventario(self):
        print("El inventario es el siguiente:") #añado un print antes de empezar con los bucles para que sea como un título del inventario
        for producto in self.inventario: #delimito que el producto tiene que estar en el inventario 
            detalles = [] #creo una lista vacía donde la función va a mostrar el inventario 
            for key,value in producto.items(): #itero el diccionario con este bucle for
                detalles.append(f"{key.capitalize()}: {value}") #le pido que añada la key y la value del inventario con un append y con capitalize para que la primera letra sea mayúscula
            print(", ".join(detalles)) #le pido que añada una coma para separar los elementos y que así quede más visual 
               

    def buscar_producto(self, nombre): 
        for b in self.inventario: #defino que tiene que estar en el inventario con el bucle for 
            if b["nombre"] == nombre: #si el nombre dado para buscar en el inventario coincide con algo que esté en el inventario
                print(f"Nombre: {b['nombre']}, Precio: {b['precio']}€, Cantidad: {b['cantidad']}") #le pido que imprima este texto 
                return #termina el bucle para que no imprima los dos prints
        b["nombre"] != nombre #si el producto no está en el diccionario, deja de buscar 
        print("El producto indicado no está en el inventario") #imprime esto para indicar que no está dentro 

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario: #defino que el producto tiene que estar en el inventario con el bucle for 
            if producto["nombre"] == nombre: #si el nombre dado para buscar en el inventario coincide con algo que esté en el inventario
                producto["cantidad"] += cantidad #indico que si el nombre coincide, que actualice la cantidad 
                print("El stock ha sido actualizado") 
                return #cierro el bucle para que no imprima ambos prints si no el que corresponda 
        print("El producto indicado no está en el inventario. Por favor, primero actualiza el inventario e inténtalo de nuevo") #ya fuera del bucle, si el producto no está en el inventario después de recorrerlo 
    
    def eliminar_producto(self, nombre):
        for producto in self.inventario: #defino que el producto tiene que estar en el inventario con el bucle for
            if producto["nombre"] == nombre: #si el nombre dado para buscar en el inventario coincide con algo que esté en el inventario
                self.inventario.remove(producto) #añado la función remove para que lo elimine de la lista del inventario
                print(f"El siguiente producto ha sido eliminado: {producto}")
                return #cierro el bucle para que no busque el siguiente producto que eliminar si no está en el diccionario 
        print("El producto indicado no está en el inventario, por favor revisa el nombre del producto e inténtalo otra vez ") #fuera del bucle, si no lo encuentra que no elimine nada 
                        
```
