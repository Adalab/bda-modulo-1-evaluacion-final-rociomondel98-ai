```python
class TiendaOnline: #definimos la clase 
    def __init__(self, nombre): #arrancamos el constructor 
        self.nombre = nombre #añado la variable de nombre para poder ponerle un nombre a la tienda 
        self.inventario = [] #añado una lista vacía que será de diccionarios 
        self.ventas_totales = 0.0 #empezamos con 0 ventas totales, con este formato porque es un float

    def agregar_producto(self, nombre, precio, cantidad):
        nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad} 
        for producto in self.inventario:
            if producto["nombre"] == nombre: 
                producto["cantidad"] += cantidad
                return
        nuevo_producto not in self.inventario
        self.inventario.append(nuevo_producto)
        print(f"Se ha añadido el producto {nombre} al inventario y hay un total de {cantidad}")

    def ver_inventario(self):
        print("El inventario es el siguiente:")
        for producto in self.inventario:
            detalles = []
            for key,value in producto.items():
                detalles.append(f"{key.capitalize()}: {value}")
            print(", ".join(detalles))
               

    def buscar_producto(self, nombre): 
        for b in self.inventario:
            if b['nombre'] == nombre:
                print(f"Nombre: {b['nombre']}, Precio: {b['precio']}€, Cantidad: {b['cantidad']}")
            else:
                print("El producto indicado no está en el inventario")
                return 
```

```python
glowy = TiendaOnline("mi tienda")
```

```python
glowy.agregar_producto("Colorete", 25, 4)
```

```python
glowy.agregar_producto("Iluminador", 30, 5)
```

```python
glowy.agregar_producto("Gloss", 10, 25)
```

```python
glowy.ver_inventario()
```
