```python
class TiendaOnline: #definimos la clase 
    def __init__(self, nombre): #arrancamos el constructor 
        self.nombre = nombre #creo el primer atributo para poder asignarle un nombre a la tienda  
        self.inventario = [] #creo el segundo atributo con el inventario de la tienda que ahora es una lista vacía pero será una lista de diccionarios con: nombre de producto, precio, cantidad 
        self.ventas_totales = 0.0 #creo el tercer atributo con un contador de ventas totales, he puesto un número decimal aunque sea cero para que sea una variable tipo float desde el principio 

    def agregar_producto(self, nombre, precio, cantidad):
        nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad} #añado la estructura que tiene que seguir para añadir un nuevo producto en un diccionario
        for producto in self.inventario: #comienzo un bucle for para que recorra cada diccionario de la lista con el inventario a ver si el producto está ya en stock o no 
            if producto["nombre"].lower() == nombre.lower(): #si el nombre introducido del producto ya está en el inventario, le digo que sume la nueva cantidad a la ya existente
                #añado lower para que si el usuario busca el nombre en minúscula Python no de error y lo entienda al buscarlo en el inventario 
                producto["cantidad"] += cantidad
                print(f"El producto {nombre.capitalize()} ya consta en nuestro inventario. Se han añadido {cantidad} unidades.")
                #uso el capitalize para que el nombre del producto salga en mayúsculas y quede más visual 
                print(f"El total actual es de: {producto['cantidad']}")
                return #termino el bucle para que si ese es el caso pare y no siga y añada el producto como nuevo 
        self.inventario.append(nuevo_producto) #uso la función append para que añada el nuevo producto con la estructura de diccionario que he definido anteriormente   
        print(f"Se ha añadido el producto {nombre.capitalize()} al inventario y hay un total de {cantidad}.") #además especifica cuál es el nuevo producto y cuanta cantidad hay de este 

    def ver_inventario(self):
        print("El inventario es el siguiente:") #añado un print antes de empezar con los bucles para que sea como un título del inventario
        print("-" * 55)
        for producto in self.inventario: #iterando cada diccionario de la lista para recorrer cada producto 
            detalles = [] #creo una lista vacía donde la función va a mostrar el inventario 
            for key,value in producto.items(): #uso el método .items para obtener la clave y el valor de cada dato 
                if key == "precio": #añado este condicional para que al imprimir el precio me añada el símbolo de euro
                    detalles.append(f"{key.capitalize()}: {value}€")
                else: #si la key no es equivalente a precio no tiene que añadir el símbolo de euro 
                    detalles.append(f"{key.capitalize()}: {value}") #le pido que añada la key y la value del inventario con un append y con capitalize para que la primera letra sea mayúscula
            print("-" * 55)
            print(", ".join(detalles)) #le pido que añada una coma para separar los elementos y que así quede más visual 
            
               
    def buscar_producto(self, nombre): 
        for b in self.inventario: #que itere cada diccionario de la lista para buscar una coincidencia de nombre 
            if b["nombre"] == nombre: #si el nombre dado para buscar en el inventario coincide con algo que esté en el inventario
                #la b es una variable temporal 
                print("Los detalles del producto buscado son los siguientes:")
                print(f"Nombre: {b['nombre'].capitalize()}, Precio: {b['precio']}€, Cantidad: {b['cantidad']}") #le pido que imprima este texto  
                return #termina el bucle para que no imprima los dos prints 
        print("El producto indicado no está en el inventario") #imprime esto para indicar que no está dentro 

    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario: #que itere cada diccionario de la lista buscando el producto que se desea actualizar 
            if producto["nombre"] == nombre: #si el nombre dado para buscar en el inventario coincide con algo que esté en el inventario
                producto["cantidad"] += cantidad #indico que si el nombre coincide, que actualice la cantidad actual del stock de ese producto 
                print(f"El stock del producto {nombre.capitalize()} ha sido actualizado") 
                print(f"El total actual es de: {producto['cantidad']}")
                return #cierro el bucle para que no imprima ambos prints si no el que corresponda 
        print("El producto indicado no está en el inventario. Por favor, primero actualiza el inventario e inténtalo de nuevo") #ya fuera del bucle, si el producto no está en el inventario después de recorrerlo 
    
    def eliminar_producto(self, nombre):
        for producto in self.inventario: #que itere cada diccionario de la lista buscando el producto que se quiere eliminar 
            if producto["nombre"] == nombre: #si el nombre dado para buscar en el inventario coincide con algo que esté en el inventario para borrarlo 
                self.inventario.remove(producto) #añado la función remove para que elimine el diccionario del producto de la lista del inventario
                print(f"El siguiente producto ha sido eliminado: {nombre.capitalize()}")
                return #cierro el bucle para que no busque el siguiente producto que eliminar si no está en el diccionario 
        print("El producto indicado no está en el inventario, por favor revisa el nombre del producto e inténtalo otra vez ") #fuera del bucle, si no lo encuentra que no lo imprima dos veces  

    def calcular_valor_inventario(self):
        print("El valor del inventario es el siguiente:") 
        print("-" * 40)
        valor_inventario = 0 #creo una variable vacía con un integer para que luego todos los precios se sumen en ella
        for producto in self.inventario: #que itere en el diccionario contando todos los productos 
            total_por_producto = producto["precio"] * producto["cantidad"] #defino la variable del precio de cada producto con una multiplicación de la cantidad de unidades que haya por el precio por unidad  
            print(f"{producto['nombre'].capitalize()}: {producto['cantidad']} unidades x {producto['precio']}€ = {total_por_producto}€")
            print("-" * 40)
            #pongo primero el nombre del producto para saber de cuál estamos hablando, después la cantidad y el precio, y luego la operación que he definido arriba
            valor_inventario += total_por_producto #le pido que sume las operaciones por cada producto y las sume al valor total acumulado del inventario 

        print(f"El VALOR TOTAL DEL INVENTARIO es: {valor_inventario}€") #imprimo el valor total del inventario                    
```
